```python
# Question
"""You are an HR Analyst. You need to build a tool that predicts a fair salary based on years of experience
```

```python
# Library Importation
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Generating Synthetic Data
np.random.seed(42)
experience = np.arange(1, 21)
salary = 30000 + (experience * 5000) + np.random.randint(-2000, 2000, size=20)
```

```python
#
df = pd.DataFrame({'Years': experience, 'Salary': salary})
print(df.head())
```

```python
# Define your X (Years) and y (Salary).
x = df[['Years']]
y = df['Salary']
```

```python
# Train a Linear Regression model.
model = LinearRegression()
model.fit(x, y)
print("Training complete!")
```

```python
# Predict the salary for a new hire with 12.5 years of experience.
new_hire = [[12.5]]
predicted_salary = model.predict(new_hire)
print(f"Predicted salary for 12.5 years of experience: ${predicted_salary[0]:.1f}")
```
